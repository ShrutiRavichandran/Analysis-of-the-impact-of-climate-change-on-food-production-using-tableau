```python
import pandas as pd
```

```python
a = pd.read_csv("Emissions_Totals_E_All_Data/Emissions_Totals_E_All_Data_NOFLAG.csv", encoding='latin-1')
```

```python
print(a.head(5))
```

```python
print(a.columns)
```

```python
for year in range(1961, 2051):
    col = 'Y' + str(year)
    if col in a.columns:
        a[col] = a[col].fillna(0)
print(a.head(5))
```

```python
emission_types = a['Element'].unique()
areas = a['Area'].unique()
store = dict()
store['Year'] = list()
store['Area'] = list()
for emission in emission_types:
    store[emission] = list()
    for year in range(1961, 2022):
        for area in areas:
            store[emission].append(a[(a['Element'] == emission) & (a['Area'] == area)]['Y'+str(year)].sum())
for year in range(1961, 2022):
    for area in areas:
        store['Year'].append(year)
        store['Area'].append(area)
df = pd.DataFrame(store)
print(df.head(10))
df.to_csv('emissions.csv', index = False)
```

```python
store = dict()
store['Year'] = list()
store['Annual Anomaly'] = list()
store['Annual Unc.'] = list()
with open('Land_and_Ocean_summary.txt') as f:
    for line in f:
        if not line.startswith("%") and len(line.strip()) != 0:
            l = line.split(" ")
            res = list()
            for ele in l:
                if ele.strip() != "":
                    res.append(ele)
            store['Year'].append(res[0])
            store['Annual Anomaly'].append(res[1])
            store['Annual Unc.'].append(res[2])
f.close()
df_temp = pd.DataFrame(store)
df_temp = df_temp.set_index('Year')
print(df_temp.head(5))
df_temp.to_csv('temperatures.csv')
```
