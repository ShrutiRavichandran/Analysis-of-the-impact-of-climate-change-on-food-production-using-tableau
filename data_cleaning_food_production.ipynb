```python
import pandas as pd
```

```python
df = pd.read_csv('Production_Crops_Livestock_E_All_Data/Production_Crops_Livestock_E_All_Data_NOFLAG.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
print(df.shape)
```

```python
print(df['Element'].unique())
```

```python
production_data = df[df['Element'] == 'Production']
production_data.info()
```

```python
item_types = production_data['Item'].unique()
areas = production_data['Area'].unique()
store = dict()
store['Year'] = list()
store['Area'] = list()
for item in item_types:
    store[item] = list()
    for year in range(1961, 2022):
        for area in areas:
            store[item].append(production_data[(production_data['Item'] == item) & (production_data['Area'] == area)]['Y'+str(year)].sum())
for year in range(1961, 2022):
    for area in areas:
        store['Year'].append(year)
        store['Area'].append(area)
df_new = pd.DataFrame(store)
print(df_new.head(10))
```

```python
df_new.info()
```

```python
# Country wise Production (tonnes)

areas = production_data['Area'].unique()
store = dict()
store['Year'] = list()
store['Area'] = list()
store['Production'] = list()
for year in range(1961, 2022):
    for area in areas:
        store['Production'].append(production_data[production_data['Area'] == area]['Y'+str(year)].sum())
for year in range(1961, 2022):
    for area in areas:
        store['Year'].append(year)
        store['Area'].append(area)
df_country_wise_production = pd.DataFrame(store)
print(df_country_wise_production.head(10))
```

```python
df_country_wise_production.info()
```

```python
# Item wise Production (kilo tonnes)

item_types = production_data['Item'].unique()
store = dict()
store['Year'] = list()
store['Item'] = list()
store['Production'] = list()
for year in range(1961, 2022):
    for item in item_types:
        store['Production'].append(production_data[production_data['Item'] == item]['Y'+str(year)].sum())
for year in range(1961, 2022):
    for item in item_types:
        store['Year'].append(year)
        store['Item'].append(item)
df_item_wise_production = pd.DataFrame(store)
df_item_wise_production['Production'] = df_item_wise_production['Production']/1000
print(df_item_wise_production.head(10))
```
